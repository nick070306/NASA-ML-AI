```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
import seaborn as sns

# Load the data
data = pd.read_csv('sensor.csv')

# Data preprocessing
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['month'] = data['timestamp'].dt.month
data['day'] = data['timestamp'].dt.day
data['minute'] = data['timestamp'].dt.minute
data['hour'] = data['timestamp'].dt.hour
data = data.drop(['timestamp', 'Unnamed: 0', 'sensor_15'], axis=1)
data = data.fillna(method='ffill')

# Split features and labels
y = data['machine_status']
x = data.drop(['machine_status'], axis=1)

# Normalize the features
for i in x.columns:
    x[i] = (x[i] - np.min(x[i])) / (np.max(x[i]) - np.min(x[i]))

# Label encoding for the target variable
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
enc = LabelEncoder()
y = enc.fit_transform(y)
y = to_categorical(y, num_classes=3)

# Select the top 25 features using SelectKBest
x_fs = SelectKBest(chi2, k=25).fit_transform(x, y)

# Split the dataset into two halves: one for training, one for testing
x_train, x_test, y_train, y_test = train_test_split(x_fs, y, test_size=0.5, stratify=y)

# Define and compile the model
import tensorflow as tf
from keras import Sequential
from tensorflow.keras.layers import *

model = Sequential()
model.add(Dense(8, input_shape=([25]), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', 
              metrics=['categorical_accuracy', 'Precision', 'Recall', 'AUC'])

# Train the model using the first half of the data
model.fit(x_train, y_train, epochs=5, batch_size=16, validation_data=(x_test, y_test))

# Evaluate the model on the training data
train_eval = model.evaluate(x_train, y_train)
print(f"Training evaluation: {train_eval}")

# Drop the machine_status column from the second half (test set)
x_test_for_prediction = x_test.copy()  # Use test data without machine_status for prediction

# Predict the status for the second half
y_pred = model.predict(x_test_for_prediction)

# Convert the predictions from one-hot encoding to class labels
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)  # Convert test labels back to class labels for comparison

# Compare predictions with actual values
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Compare predictions with actual values
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=1)
recall = recall_score(y_test_classes, y_pred_classes, average='weighted')
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
classification_rep = classification_report(y_test_classes, y_pred_classes, zero_division=1)

# Print the results
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)


# Print the results
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```
